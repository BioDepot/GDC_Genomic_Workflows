```r
# TCGA-AB-2821 LAML sample compare
v15.data <- read.csv(file = "v15counts.csv", header = T)
v32.data <- read.csv(file = "v32counts.csv", header = T)

# remove decimals and numbers after decimals
v15.data$genes <- sub("\\..*", "", v15.data$genes)
v32.data$genes <- sub("\\..*", "", v32.data$genes)
```

```r
combined <- inner_join(v15.data, v32.data, by =  "genes")
combined <- combined %>% distinct(genes, .keep_all = T)
combined
```

```r
write.csv(combined, file = 'GDC TCGA-AB-2821 published counts compare.csv', row.names = F)
```

```r
# Function to get histograms of relative change for 2 specified columns (indices) in one dataframe
relativeChange.hist <- function(df, original, new) {
    difference <- vector()
    tozero <- vector()

    name <- paste(colnames(df)[original], " to ", colnames(df)[new])

    for (k in 1:nrow(df)) {
        if (df[k, original] == 0) { # "divide by zero" results in a seperate vector
            tozero <- append(tozero, df[k, new])
        } else {
            # relative change value calculated and stored in vector
            change <- abs(df[k, new] - df[k, original]) / df[k, original]
            difference <- append(difference, change)
        }
    }
#     print(paste0("# Genes with original (v15) = 0: ", length(tozero)))
#     print(paste0("# Genes with change < 0.05: ", length(difference[difference < 0.05])))
#     print(paste0("# Genes with change >= 0.05: ", length(difference[difference >= 0.05])))
    
    # Histogram for relative change "difference"
    hist(difference, main = name, xlab = name, breaks = 100)

    # Histogram for change (from 0 to some non-zero value, "divide by zero" results)
    if (length(tozero) != 0) {
        hist(tozero, main = paste(name, "original=0"), xlab = paste(name, "original=0"), breaks = 100)
    } 
}
```

```r
relativeChange.hist(combined, 2, 3)
```
